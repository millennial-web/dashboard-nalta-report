```python
# cargar dependencias
import pandas as pd
import geopandas as gp
```

```python
# cargar poligonos municipales de Mexico
poligonos = gp.read_file("poligonos-mex/conjunto_de_datos/00mun.shp")
poligonos.head()
```

```python
#cargar datos censo
datos_censo = pd.read_csv("censos-conteos-2010/conjunto_de_datos/iter_00_cpv2010.csv")
#filtrar datos por Total del Municipio
datos_municipales = datos_censo[datos_censo.nom_loc == "Total del Municipio"]
datos_municipales.head()
```

```python
#format number ids to match zero prepended strings in poligons data
datos_municipales["CVE_ENT"] = datos_municipales.entidad.astype(str).str.zfill(2)
datos_municipales["CVE_MUN"] = datos_municipales.entidad.astype(str).str.zfill(3)
```

```python
datos_municipales
```

```python
#explorar los datos que podemos mapear 
# vph_autom = Viviendas particulares habitadas que tienen automóvil o camioneta
datos_municipales['vph_autom']
```

```python
#clean rows with null vals in vph_autom or nom_mun cols
datos_municipales = datos_municipales.dropna(subset=['vph_autom'])
datos_municipales
```

```python
# Hacer merge con los datos de poligonos (manteniendo el objeto en formato geopandas)
municipios_automovil = poligonos.merge(datos_municipales[["CVE_ENT","CVE_MUN","nom_mun","vph_autom"]], on = ["CVE_ENT","CVE_MUN"], how = "left")
```

```python
#revisar proyección usada por el INEGI
municipios_automovil.crs
```

```python
# convertir a la proyección estandar para la web EPSG:4326
municipios_automovil = municipios_automovil.to_crs("EPSG:4326")
```

```python
# confirmar que se cambió el formato de proyección a EPSG:4326
municipios_automovil.crs
```

```python
#verificar los nuevos parametros de Lat Lng en los poligonos
municipios_automovil.head()
```

```python
#simplify geometry of polygons to lower filesize and processing time
municipios_automovil['geometry'] = municipios_automovil['geometry'].simplify(tolerance=0.001,preserve_topology=True)
```

```python
#save the new geopandas dataframe as GeoJSON
municipios_automovil.to_file("dashboard-app/frontend-d3-leaftlet-app/assets/data/municipios_automovil.geojson", driver='GeoJSON')
```
